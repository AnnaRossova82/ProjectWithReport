```python
# For project import datset from Kaggle (platform for analytics and predictive modeling competitions)
# https://www.kaggle.com/datasets/prachi15gupta98/airline-passenger-satisfaction
data=pd.read_excel('AirlinePassengerSatisfaction.xlsx', index_col=0) 
data.head()
```

```python
#First let us check the type of objects in all columns and all data information
data.info()
```

```python
#We can see that some data is missing and we need to fill it, and also transform object data into int
#See statistics for all columns
data.describe(include = "all")
```

```python
#Check if there are any duplicates, see that there are not
data[data.duplicated()]
```

```python
#as Satisfaction is my target variable, let's check how many unique type of satisfaction is there in the dataset
data['satisfaction'].value_counts()
```

```python
# Transform four text columns into int meanings
labelencoder = LabelEncoder()
data['Customer Type'] = labelencoder.fit_transform(data['Customer Type'])
data['Gender'] = labelencoder.fit_transform(data['Gender'])
data['Type of Travel'] = labelencoder.fit_transform(data['Type of Travel'])
data['Class'] = labelencoder.fit_transform(data['Class'])
```

```python
#Now we see that all text data is transformed into int meanings (Gender,Customer Type, Type of travel, Class columns)
data.head()
```

```python
#Now check if there are missing data values, we can see that there are some (shown as True values):
missing_data = data.isnull()
missing_data.head(15)
```

```python
#So now check the proportion of Missing Values to all columns data:
nulls = data.isnull().sum()/len(data)
nulls[nulls.values > 0]
```

```python
#Check the quantity of missing values:
data.isnull().sum()
```

```python
#Compare Null and filled 'Arrival Delay in Minutes' quantity, see that it is almost equal
data.groupby(data['Arrival Delay in Minutes'].isnull()).mean()
```

```python
# Create a new column 'Delay_NaN'
# If there is no value in the 'Arrival Delay in Minutes' column, then = 0 and yes value = 1
data['Delay_NaN'] = np.where(data['Arrival Delay in Minutes'].isnull(), 0,1)
data.head(6)
```

```python
#Now check satisfaction depend on delay statistisc:
data.groupby(['Delay_NaN']) ['satisfaction'].value_counts(normalize=True)
```

```python
#Let us check correlation of'Departure Delay in Minutes' and 'Arrival Delay in Minutes' to see what to do with this column
stats.pearsonr(data.dropna()['Departure Delay in Minutes'], data.dropna()['Arrival Delay in Minutes'])
```

```python
#Now after investigating the matter we can drop not needed column we made only for checking
data.drop('Delay_NaN', axis= 1 , inplace= True )
```

```python
#As we see missing 393 numerals in 'Arrival Delay in Minutes', fill by 'Departure Delay in Minutes' as much correlated 0.9656
data['Arrival Delay in Minutes'].fillna(data['Departure Delay in Minutes'], inplace=True) # filled the missing values
```

```python
# quantity of other missing data is not significant, so replace missing data for mean:
data = data.fillna(data.mean ())
```
